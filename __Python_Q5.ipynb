# Question 5 :
Write a program to download the data from the given API link and then extract the following data with proper formatting.

Link - http://api.tvmaze.com/singlesearch/shows?q=westworld&embed=episodes

```python
import json
from urllib.request import urlopen

def get_formatted_json(url:str)->json.dumps:
    '''
    Scrapes json from a URI and returns a formatted json object
    params:
        url: (str) URL of the json resource
    returns:
        formatted json
    '''
    raw_data = urlopen(url).read()                         # scrapping raw data from the given url and reading it
    
    encoded_data = raw_data.decode('utf-8')                # encoding the byte stream to utf-8 encoding
    
    parsed_json = json.loads(encoded_data)                 # parsing the encoded charachter stream using json parser
    
    formatted_json = json.dumps(parsed_json, indent = 5)   # formatting the parsed_json
    
    return formatted_json

if __name__ == "__main__":
    url = 'http://api.tvmaze.com/singlesearch/shows?q=westworld&embed=episodes'
    
    output = get_formatted_json(url)
    
    print(output)
```
